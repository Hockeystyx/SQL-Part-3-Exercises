> <span style="font-family: -apple-system, BlinkMacSystemFont, sans-serif; background-color: rgba(127, 127, 127, 0.1);"><i>Minimum Desired Output:</i>&nbsp;The tag name for the most popular tag used for the most tagged book.&nbsp;</span>    
> 
> <span style="font-family: -apple-system, BlinkMacSystemFont, sans-serif; background-color: rgba(127, 127, 127, 0.1);"><b>Tag Name: To Read list</b></span>

```sql
-- Code Here:
SELECT DISTINCT b.title, t.count, b.best_book_id, t.tag_id
FROM BooksDB.dbo.books AS b
INNER JOIN BooksDB.dbo.book_tags AS t
ON b.best_book_id = t.goodreads_book_id
WHERE b.best_book_id = 865
ORDER BY t.count DESC;

SELECT tag_name FROM BooksDB.dbo.tags
WHERE tag_id = 30574;
```
